```python
import pandas as pd
import cufflinks as cf
import numpy as np
close_data_aomeiyiliao=df1[['trade_date','close']]
close_data_yangpuyiliao=df2[['trade_date','close']]
close_data_taidagufen=df3[['trade_date','close']]
close_data_aomeiyiliao.columns=['trade_date','AoMei']
close_data_yangpuyiliao.columns=['trade_date','Yangpu']
close_data_taidagufen.columns=['trade_date','TaiDa']
close_data_middle=pd.merge(close_data_aomeiyiliao,close_data_yangpuyiliao,on="trade_date")
close_data_final=pd.merge(close_data_middle,close_data_taidagufen,on ="trade_date")
close_data_final.trade_date=pd.to_datetime(close_data_final.trade_date)
close_data_final.set_index(["trade_date"], inplace=True)
# #close_data_final

close_data_final.iplot(theme='pearl', filename='Tutorial Metadata', title='Chinese Mask Companies Stock Price',xTitle='Date', yTitle='Stock Price', world_readable=True)
```
